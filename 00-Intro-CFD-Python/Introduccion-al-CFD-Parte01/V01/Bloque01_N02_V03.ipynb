```python
%pylab inline

import numpy as np
import matplotlib.pyplot as plt

def linearconv(nx):
    dx = 2./(nx-1)
    nt = 20
    dt = .025
    c = 1
    
    u = np.ones(nx)
    u[int(.5/dx) : int(1/dx+1)] = 2
    
    un = np.ones(nx)
    
    for n in range(nt):
        un[:] = u[:]
        for i in range(1, nx):
            u[i] = un[i]-c*dt/dx*(un[i]-un[i-1])
            
    plt.plot(np.linspace(0,2,nx),u)
```

```python
linearconv(41)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def linearconv(nx):
    dx = 2./(nx-1)
    nt = 20
    c = 1
    sigma = .5
    
    dt = sigma*dx
    
    u = np.ones(nx)
    #u[.5/dx : 1/dx+1]=2
    u[int(.5/dx) : int(1./dx+1)]=2

    
    un = np.ones(nx)
    
    for n in range(nt):
        un[:] = u[:]
        for i in range(1, nx):
            u[i] = un[i]-c*dt/dx*(un[i]-un[i-1])
    plt.plot(np.linspace(0,2,nx),u)
```
